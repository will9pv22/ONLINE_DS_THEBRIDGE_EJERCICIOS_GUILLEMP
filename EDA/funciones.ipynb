```python
def categorize_income(pib_per_capita):
    if pib_per_capita > 13205:
        return 'Ingresos Altos'
    elif 1046 <= pib_per_capita <= 4256:
        return 'Ingresos Medios'
    else:
        return 'Ingresos Bajos'

df_felicidad['Income Category'] = df_felicidad['PIB per capita'].apply(categorize_income)

# Mostrar el DataFrame con la nueva columna
df_felicidad
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

def tabla_correlacion(country, year):
    # Filtrar los datos por paÃ­s y aÃ±o
    filtered_data = df_felicidad[(df_felicidad['Country name'] == country) & (df_felicidad['year'] == year)]
    
    if filtered_data.empty:
        return "No data available for the specified country and year."
    
    # Seleccionar las columnas relevantes
    relevant_columns = [
        'Life Ladder', 'Log GDP per capita', 'Social support',
        'Healthy life expectancy at birth', 'Freedom to make life choices',
        'Generosity', 'Perceptions of corruption', 'Positive affect', 'Negative affect'
    ]
    
    filtered_data = filtered_data[relevant_columns]
    
    # Calcular la tabla de correlaciÃ³n
    correlation_matrix = filtered_data.corr()
    return correlation_matrix
```
